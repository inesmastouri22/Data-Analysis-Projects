```python
ax = df['Year_Introduced'].value_counts() \
    .head(20) \
    .plot(kind='bar', title='Top 20 Years Coasters Introduced', color='#779ECB')  # Light pastel blue

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')

plt.show()
```
